```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset1 = pd.read_csv("distdata.csv",header=None).fillna(0)
```

```python
n,m=dataset1.shape
place = ["Front","Front Right","Front Left","Behind Right","Back"]
lists = []

for i in range(n):
    for j in range(m):
        lists.append(dataset1[j][i])
        #
        #3n-2,3n-13n

lists = np.array(lists)
#

mat = lists.reshape([n,m])
#m*n

# print(mat)
#

plt.figure(figsize=(9,6))

for i in range(n):
    plt.plot(mat[i,:])
    #

plt.legend(place)
plt.xlabel("Time")
plt.ylabel("Distance")

plt.show()
```
