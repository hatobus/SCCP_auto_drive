```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset1 = pd.read_csv("distdata000.csv",header=None).fillna(0)
```

```python
n,m=dataset1.shape
place = ["Front","Front Right","Front Left","Behind Right","Back"]
lists = []

for i in range(n):
    for j in range(m):
        lists.append(dataset1[j][i])
        #リスト内にデータを入れていく．
        #O(n*m)なので圧倒的に速度が遅いと思われるのでここはあとで書き直しておく．

#後でリスト内包表記に変えておく．
        
lists = np.array(lists)
#ここでリスト内を平坦化

mat = lists.reshape([n,m])
#これでm*nの行列が生成される．

# print(mat)
#一応プリントしておく

plt.figure(figsize=(9,6))

for i in range(n):
    plt.plot(mat[i,:])
    #スライシングとループ処理を使用して，一列ごとにプロット．

plt.legend(place)
plt.xlabel("Time")
plt.ylabel("Distance")

plt.show()
```
