# データをプロットする処理
データを取得し，処理してからグラフにプロットするファイル．

やっていることとしてはcsvファイルからデータを読み込み，それを行列にする．

それをスライングで情報を取得することにより，データ数が増えても対応が可能な処理

```python
#coding:utf-8
import numpy as np
import matplotlib.pyplot as plt
```

```python
#coding:utf-8
import numpy as np
import matplotlib.pyplot as plt

x = np.loadtxt('distdata.csv', delimiter=',')
n = x[0].size
m = np.ndim(x)+1
print(m)

place = ["Front","Front Left","Front Right"]
lists = []

for i in range(n):
    for j in range(m):
        lists.append(x[j][i])
        #リスト内にデータを入れていく．
        #これだと3n-2のデータがフロントで,3n-1が左前，3nが右前になる

lists = np.array(lists)
#ここでリスト内を平坦化

mat = lists.reshape([n,m])
#これでm*nの行列が生成される．

print(mat)
#一応プリントしておく

for i in range(m):
    plt.plot(mat[:,i], label=str(i))
    #スライシングとループ処理を使用して，一列ごとにプロット．

plt.legend(place)
plt.xlabel("Time")
plt.ylabel("Distance")

plt.show()
```
