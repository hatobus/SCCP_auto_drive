# データをプロットする処理
データを取得し，処理してからグラフにプロットするファイル．

やっていることとしてはcsvファイルからデータを読み込み，それを行列にする．

それをスライングで情報を取得することにより，データ数が増えても対応が可能な処理

```python
#coding:utf-8
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.loadtxt('distdata.csv', delimiter=',')
n = x[0].size
m = np.ndim(x)+1
print(m)

lists = []

for i in range(n):
    for j in range(m):
        lists.append(x[j][i])

lists = np.array(lists)
# print(lists)
mat = lists.reshape([n,m])
print(mat)

for i in range(m):
    plt.plot(mat[:,i])

plt.show()
```
