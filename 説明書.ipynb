# Jupyer notebookの使い方



Jupyter notebookはオンライン上で自分のPCにアクセス出来、実際にコードもかけるというツールです

```python
# -*- coding: utf-8 -*- 
print("Jupyter notebook!")
```

上のようにプログラムを書いて、実行（Shift + Enter）をすると、実際にプログラムを動かすことができます。

また、今書いているようにMarkdownを使用して、今何をしているかというのもわかりやすいようにコメントを残すこともできます。

また、Pythonのコードなら何でも書けるので、ライブラリなどをimportすることで、いろいろなことができます。

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*- 

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(-3, 3, 0.1)
y = np.sin(x)
plt.plot(x,y)
```

このようにmatplotlibを利用して、グラフを描画することができます。
また、画像を表示することなども可能です。

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*- 
from matplotlib.image import imread

img = imread('aidu.png')

plt.imshow(img)
plt.show()
```
